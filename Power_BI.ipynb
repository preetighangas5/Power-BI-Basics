**POWER BI BASICS**

**Question 1:** What is Power BI and why is it used in business ?

**Answer:** Power BI is a Business Intelligence and data visualization tool developed by Microsoft. It allows users to connect to various data sources, transform and clean data, create interactive reports and dashboards, and share insights across an organization.

Why is it used in business:

-  Helps turn raw data into meaningful insights through interactive visualizations.
-  Enables data-driven decision making.
-  Supports real-time monitoring of KPIs and business performance.
-  Reduces dependency on IT teams by empowering business users  (self-service BI).
-  Integrates well with other Microsoft tools like Excel, Azure, Teams, and SharePoint.
-  Cost-effective and scalable for organizations of all sizes.


**Question 2:** Name & explain three main components of Power BI.

**Answer:** The three main components of Power BI are:

1. *Power BI Dekstop*

Free Windows application used for creating reports. It is the primary authoring tool where you connect to data, transform it (Power Query), build data models (relationships, DAX), and design reports.

2. *Power BI Service (PowerBI.com)*

Cloud-based SaaS platform. You publish reports created in Desktop here. It allows sharing, collaboration, scheduling data refreshes, creating dashboards, and viewing reports on the web or mobile.

3. *Power BI Mobile*

Apps for Windows, iOS, and Android that allow users to view and interact with reports and dashboards on mobile devices.

**Question 3:** Explain the Power BI Workflow.

**Answer:** The typical Power BI Workflow consists of the following steps:

1. Connect & Extract - Connect to various data sources (Excel, SQL, Web, etc.).
2. Transform & Clean - Use Power Query Editor to clean, shape, and transform data.
3. Model - Build relationships between tables and create calculated measures/columns using DAX.
4. Visualize - Create interactive reports and dashboards in Report View.
5. Publish - Publish the report to the Power BI Service.
6. Share & Collaborate - Share reports/dashboards with others and set up scheduled refresh.
7. Consume - View reports on web, mobile, or embed them in other applications.


**Question 4:** List any four data cleaning tasks that can be performed in Power Query.

**Answer:** For Commn Data Cleaning tasks in Power Query:

1.  Remove Duplicates - Eliminate duplicate rows.
2.  Handle Missing Values -- Replace nulls with specific values, remove rows with nulls, or fill down/up.
3.  Change Data Types - Convert columns to correct data types (Text, Number, Date, etc.).
4.  Split Columns - Split one column into multiple columns (by delimiter, number of characters, etc.).
5.  Trim & Clean - Remove extra spaces or non-printable characters.
6.  Filter Rows - Keep or remove rows based on conditions.

**Question 6:**  Define Data View, Report view and Model view.Explain the purpose of each view.

**Answer:** Power BI Dekstop has three main views that help you work with your data at different stages.

-  ***Data View***

Allows you to explore the actual data in tabular format. You can see all rows and columns, add calculated columns, and inspect data quality. Best for viewing raw or transformed data.

PURPOSE:

1. Browse and inspect the actual rows and columns of your tables.
2. Check data quality and values.
3. Create calculated columns using DAX.
4. Format columns (change data types, add currency symbols, set decimal places, etc.).
5. Apply sorting or filtering to explore the data.

-  ***Report View***

Default view. Used to create interactive visualizations (charts, tables, maps, slicers, etc.) and design the final report layout for end users.

PURPOSE:

1. Create interactive visualizations (charts, graphs, tables, maps, slicers, cards, etc.).
2. Design the layout and appearance of the report for end users.
3. Add filters, slicers, buttons, and bookmarks to make the report interactive.
4. Arrange visuals to tell a story with the data.

-  ***Model View***

Shows all tables and their relationships (like a database diagram). Used to create, edit, and manage relationships between tables, hide columns, and optimize the data model.

PURPOSE:

1. Create and manage relationships between different tables (e.g., linking Sales table with Date table or Property table with Owner table).
2. View the entire data model like a diagram.
3. Hide unnecessary columns from end users.
4. Optimize the model for better performance.
5. Define hierarchies and set up cross-filtering behavior.

**Question 7:** Discuss the different Data Sources that Power BI supports.

**Answer:** Power BI supports a variety of data sources, categorized as:

- *Files:*  Excel, CSV, Text, JSON, XML, PDF.
- *Databases:*  SQL Server, Oracle, MySQL, PostgreSQL, Snowflake, etc.
- *Cloud Services:*  Azure SQL, Azure Data Lake, Google BigQuery, Amazon Redshift.
- *Online Services:*  SharePoint, Dynamics 365, Salesforce, Google Analytics, Facebook, etc.
- *Web:*  Connect directly to web pages or APIs.
- *Other:*  OData, ODBC, OLE DB, Power BI datasets, Dataverse, etc.


**Question 8:** Split Owner Name to create two New columns as First Name and Last Name.

**Answer:** Steps in Power Query Editor:

1. Open the query in Power Query Editor.
2. Select the "Owner Name" column.
3. Go to the "Transform" tab → Click "Split Column" → Choose "By Delimiter".
4. Select the delimiter as Space.
5. Choose "Split at: Left-most delimiter" (for First and Last Name).
Rename the new columns to "First Name" and "Last Name".
6. Click "Close & Apply".
